```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import os
import glob
```

```python
path = "DATASET/*.txt"
files = glob.glob(path)
# happy, sad, sar, angry = [], [], [], []
for name in files:
    with open(name) as f:
        data = f.read()
    
        if name == 'DATASET/happy.txt':
            happy = data.split('\n')
        if name == 'DATASET/sad.txt':
            sad = data.split('\n')
        if name == 'DATASET/sarcastic.txt':
            sar = data.split('\n')
        if name == 'DATASET/angry.txt':
            angry = data.split('\n')
```

```python
happy_set = pd.DataFrame()
happy_set['text'] = happy
happy_set['sentiment'] = 'happy'

sad_set = pd.DataFrame()
sad_set['text'] = sad
sad_set['sentiment'] = 'sad'

sar_set = pd.DataFrame()
sar_set['text'] = sar
sar_set['sentiment'] = 'sar'

angry_set = pd.DataFrame()
angry_set['text'] = angry
angry_set['sentiment'] = 'angry'
```

```python
dataset = pd.concat([happy_set, sad_set, sar_set, angry_set], ignore_index=True)
dataset = dataset.sample(frac=1).reset_index(drop=True)

dataset.head(100)
```

```python
print(dataset.describe())
```

```python
stat_by_sent = dataset['sentiment'].value_counts()
fig, ax = plt.subplots()
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=15)
ax.set_ylabel('Number of statements' , fontsize=15)
ax.set_title('SENTIMENTS', fontsize=10, fontweight='bold')
stat_by_sent.plot(ax=ax, kind='bar', color='blue')
plt.show()
```
